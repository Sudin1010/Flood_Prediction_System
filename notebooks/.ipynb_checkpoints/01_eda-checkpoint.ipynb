# 01 EDA — Flood Prediction System

Run all cells. This notebook shows dataset understanding, missing values, distributions, and correlation analysis.

```python
from pathlib import Path
import sys

CURRENT_DIR = Path.cwd()
PROJECT_ROOT = CURRENT_DIR.parent if CURRENT_DIR.name == "notebooks" else CURRENT_DIR
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print("Project Root:", PROJECT_ROOT)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_PATH = PROJECT_ROOT / "data" / "flood_dataset_clean.csv"
df = pd.read_csv(DATA_PATH)

print("Dataset loaded successfully")
print("Rows and columns:", df.shape)
df.head()
```

```python
print("Column names:")
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {col}")
```

```python
df.info()
```

```python
missing = df.isnull().sum()
print("Missing values:")
missing
```

```python
df.describe().T
```

```python
df["FloodRisk"] = (df["FloodProbability"] >= 0.5).astype(int)
print("Target converted: FloodProbability >= 0.5 means FloodRisk = 1")
df[["FloodProbability", "FloodRisk"]].head(10)
```

```python
risk_counts = df["FloodRisk"].value_counts().sort_index()
risk_counts.index = ["No Flood Risk (0)", "Flood Risk (1)"]
risk_counts
```

```python
plt.figure(figsize=(6,4))
df["FloodProbability"].hist(bins=30)
plt.title("Distribution of Flood Probability")
plt.xlabel("Flood Probability")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
df["FloodRisk"].value_counts().sort_index().plot(kind="bar")
plt.title("Flood Risk Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
corr = df.drop(columns=["FloodRisk"]).corr(numeric_only=True)["FloodProbability"].sort_values(ascending=False)
corr
```

```python
plt.figure(figsize=(9,7))
corr_matrix = df.drop(columns=["FloodRisk"]).corr(numeric_only=True)
plt.imshow(corr_matrix, aspect="auto")
plt.colorbar(label="Correlation")
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.index)), corr_matrix.index)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```

## EDA Conclusion

The dataset is clean and all features are numerical. For mid-defense, `FloodProbability` is converted to `FloodRisk` so the system can use a from-scratch classification algorithm.
